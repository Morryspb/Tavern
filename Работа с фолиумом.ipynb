```python
import pandas as pd
import folium
import numpy as np
```

```python
# Import folium MarkerCluster plugin
from folium.plugins import MarkerCluster
# Import folium MousePosition plugin
from folium.plugins import MousePosition
# Import folium DivIcon plugin
from folium.features import DivIcon
```

```python
# define the world map
world_map = folium.Map()

# display world map
world_map
```

```python
cyprus_map = folium.Map(
    location = [34.7720133, 32.4297369],    # широта и долгота России
    zoom_start = 10
)

# display russia map
cyprus_map
```

```python
data=pd.read_excel('geo_fin.xlsx')
```

```python
data
```

```python
for idx, row in data.iterrows():
    rating = row['rating']
    if rating < 9:
        color = 'red'
    elif 9 <= rating <= 9.3:
        color = 'orange'
    else:
        color = 'green'
        
    popup_html = f"""
    <span style='font-size:12pt; font-weight:bold'>{row['name']}</span>
    <br>Price: {row['price']}
    <br>Rating: {row['rating']}
    <br>Особенности: {row['fiches']}
    """
    # Если location - строка "lat, lon"
    lat, lon = map(float, row['location'].split(','))
    folium.Marker(
        location=[lat, lon],
        popup=folium.Popup(popup_html, max_width=250),
        icon=folium.Icon(color=color)
    ).add_to(cyprus_map)
```
